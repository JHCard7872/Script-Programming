- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> c = c + 5\
>\>>> d = d - 5\
>\>>> print(c)\

>\>>> print(d)\
[Count (step: 2)] 9

- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> print(c) > 10\
True\
>\>>> print(d) > 10\
False\
>\>>> print(c) < 10\
False\
>\>>> print(d) < 10\
True\
>\>>> print(c) == 17\
True\
>\>>> print(d) != 9\
False

```python
class Counter:
    def __init__(self, count, step=1): #요구사항 1
        self.count = count
        self.step = step
        
    def __str__(self): # 요구사항 2
        return "[Count(step: " + str(self.step) + ")] " + str(self.count)
    
    def incr(self): # 요구사항 3
        self.count += self.step
        
    def __call__(self): # 요구사항 4
        self.incr()
        
    def __add__(self, other): # 요구사항 5
        self.count += other
        return self
        
    def __sub__(self, other):
        self.count -= other
        return self
    
    def __eq__(self, other): # 요구사항 6
        return self.count == other

    def __lt__(self, other):
        return self.count < other

    def __gt__(self, other):
        return self.count > other
         
c = Counter(10) # 요구사항 1
d = Counter(10, 2)

print(c) # 요구사항 2
print(d)

c.incr() # 요구사항 3
d.incr()

print(c)
print(d)

c() # 요구사항 4
d()
print(c)
print(d)

c = c+5 # 요구사항 5
d = d-5
print(c)
print(d)

print(c>10) # 요구사항 6
print(d>10)
print(c<10)
print(d<10)
print(c==17)
print(d != 9)
```

각 요구사항에 맞게 적절한 메소드들을 추가해주어 각 역할에 맞게 코딩을 해주었다.

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

\__init\__ - 생성자이다. s = MySet([1,2,2,3]) 에서 MySet의 인스턴스를 생성할때 호출이 되는 함수로 MySet의 인스턴스 s에 append를 해주는 코드가 있다. self는 s, l은 MySet과 함께 매개변수로 주는 [1,2,2,3]이 된다. 이 값을 반복문으로 s에 append를 해준뒤, MySet.eliminate_duplicate(self)함수를 호출시킨다.\
\__str\__ - print나 str() 호출에 대응되는 메소드로 "MySet: {, , , }"식으로 출력이 되게끔 하게 만드는 함수이다. 반복문으로 self에 append되어 있는 값들을 반복문으로 가져와 result에 저장해 준 뒤 이 값을 리턴해주어 출력해준다.\
eliminate_duplicate - 중복되는 값들을 제거해주는 역할을 한다. 반복문 안에 저장할 값을 조건문으로 이미 있는 값은 append하지 않는 식으로 하여 저장해 준다.

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

>\>>> u = s | t\
>\>>> print(u)\
>MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

>\>>> u = s & t\
>\>>> print(u)\
>MySet: {2, 3}

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

>\>>> s = MySet([1, 2, 3])\
>\>>> t = MySet([3, 4, 5])\
>\>>> u = s - t\
>\>>> print(u)\
>MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other): # 요구사항 1
        s = MySet([]) # 반환할 s인스턴스 생성
        for e in self: # self의 값 전부 넣어주고
            s.append(e)
        for e in other: #  이미 넣은 self에는 없는 other값 append
            if e not in self:
                s.append(e)
        return s
    
    def __and__(self, other): # 요구사항 2
        s = MySet([]) # 반환할 s인스턴스 생성
        for e in other: # 겹치는거 append
            if e in self:
                s.append(e)
        return s

    def __sub__(self, other): # 요구사항 3
        s = MySet([])
        for e in self: # self에서 other에 있는거 제외하고 전부 append
            if e not in other:
                s.append(e)
        return s
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    u = s|t  # 요구사항 1
    print(u)
    u = s&t # 요구사항 2
    print(u)
    s = MySet([1,2,3]) # 요구사항 3
    t = MySet([3,4,5]) 
    u = s - t
    print(u)
```

| 연산은 \__or\__, & 연산은 \__and\__, -연산은 \__sub\__ 이므로 각각의 함수에 맞게 수정해 주었다. MySet()인스턴스를 함수 내에서 선언해주고, s와 t를 받아 연산을 하면 s가 self에 t가 other에 들어가 연산을 시작한다. |연산은 s에 있는 값을 모두 새로 생성한 MySet인스턴스에 저장해주고, s에 없는 요소의 t의 요소도 저장해주었다. &연산은 s와 t를 비교해 서로 있는 요소만 저장해주도록 하였고, - 연산은 self를 기준으로 other와 동일한 값은 제외해주도록 하였다.
