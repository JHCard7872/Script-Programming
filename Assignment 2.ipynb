### 2. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - div_path('/usr/local/bin/python')
        - ['/usr/local/bin', 'python']
    - div_path('/home/chulsoo/test.txt')
        - ['/home/chulsoo', 'test.txt']
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

```python
def div_path(s):
    count = 0
    for ch in range(len(s)): # 매개변수의 s의 길이만큼 반복
        if(s[len(s)-ch-1]=='/'): # s배열의 거꾸로 시작해서 '/' 확인시 종료
            break
        count = count+1 # '/'를 확인을 못하면 count 증가
    dir = s[0:len(s)-count-1] # dir의 값에 s[0] 부터 s[s의 길이 - count]값 전체 저장
    file = s[len(s)-count:len(s)] # 거꾸로 생각해서 '/'부터 끝까지 저장
    print("['" + dir + "', '" + file + "']") # 출력
div_path('/usr/local/bin/python')
div_path('/home/chulsoo/test.txt')
```

문자열을 받으면 그 값을 거꾸로 확인하여 '/'값이 나올때까지 count변수를 증가 시켜준다. 그러면 count값이 파일명의 길이가 되므로 s[0:문자열의 길이 - 파일명의 길이 - 1]이 디렉토리 경로명이 되고 s[문자열의 길이 - 파일명의 길이 : 문자열의 길이]가 파일명이 된다. 이를 출력하면 결과값이 나오게 된다.

### 3. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
- 집합 자료형 set은 사용하지 않는다.
- 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
- 반환하는 리스트는 정렬이 되어 있어야 한다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - list_union([1, 2, 3], [1, 2, 4])
        - [1, 2, 3, 4]
    - list_union([-10, -5, 0, -1], [100, 9, 0, 9])
        - [-10, -5, -1, 0, 9, 100]
    - list_union([0, 1, 2], [0, 1, 2])
        - [0, 1, 2]
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
- [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb):
    list = lista + listb # 리스트에 두 개의 리스트를 저장
    list.sort() # 정렬
    count = 0
    for ch in range(len(list)): # 길이만큼 반복
        if(count==len(list)-1): # count값은 중복되는것이 있을때마다 -1, 반복 할떄마다 +1을 해줌
            break
        if(list[count] == list[count+1]): #중복되는 값이 있을때 삭제 및 count -1
            list.remove(list[count])
            count = count - 1
        count = count+1
    return list # 반환
    
print(list_union([1, 2, 3], [1, 2, 4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))
```

전체적으로 하나의 리스트 안에 2개의 리스트를 넣어주고 정렬을 한다. 중복되는 값을 찾기 위해 반복문을 실행하고 중복될 경우 그 값을 삭제한다. 삭제할 경우 리스트들이 삭제 된 자리를 메꾸는데 그것을 count를 선언해주어 해결해주었다. 삭제할때마다 count값을 -1해주고, 검사를 진행할때 마다 +1을 해준다. 리스트의 길이만큼 진행을 해주어야 하기 때문에 길이에 닿으면 break를 해주어 함수를 종료해주었다.

### 4. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 '차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.
- 모든 가정과 조건은 3번 문제와 동일하다.

```python
def list_intersection(lista, listb):
    list = []
    for ch in range(len(lista)): # lista만큼 반복
        for ch2 in range(len(listb)): #listb만큼 반복
            if(lista[ch] in listb ): #lista의 ch번째의 수가 listb에 있을경우
                if(lista[ch] in list): # 이미 list에 존재하면 continue
                    continue 
                else: # 해당할 경우 list에 저장
                    list.append(lista[ch])
    return list
print(list_intersection([1, 2, 3], [1, 2, 4]))
print(list_intersection([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_intersection([0, 1, 2], [0, 1, 2]))

def list_difference(lista, listb):
    list = []
                            
    for ch3 in range(len(listb)):
        if(listb[ch3] in lista ): # 리스트에 있을경우 continue
            continue
        else: #상대 리스트에도 없고 출력해줄 list에도 없으면 저장 있으면 continue
            if(listb[ch3] in list):
                continue
            list.append(listb[ch3])
                        
    for ch in range(len(lista)): # 나머지 리스트도 똑같이 진행
        if(lista[ch] in listb ):
            continue
        else:
            if(lista[ch] in list):
                continue
            list.append(lista[ch])
    return list
print(list_difference([1, 2, 3], [1, 2, 4]))
print(list_difference([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_difference([0, 1, 2], [0, 1, 2]))
```

intersection함수는 하나의 리스트를 선택하여 상대 리스와 반복문으로 비교하여 있을경우 새로운 list에 저장해준다. 
difference함수는 intersecion에서 중복되는 것을 확인하면 continue, 중복되지 않으면 list에 저장, 만약 이미 list에 있는 수면 continue를 해주어 list에 저장을 해주고 출력해주었다.

### 5. 두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.
- 이미 구현된 모듈이나 라이브러리는 사용하지 않는다.
- 함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.
    - print_primenumber(2, 10)
        - 2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

```python
def print_primenumber(a, b):
    list = []
    for ch in range(a, b+1): # a부터 b까지 판단해야 하므로
        check = True #소수 판단 변수
        for prime in range(2, ch):
            if(ch % prime == 0): # b-1까지의 값으로 나누어 지면 false체크
                check = False
                break
        if(check): # check가 true이면 소수
            list.append(ch)
    print('{} 이상 {} 이하 정수 중 소수 리스트: {}'.format(a, b, list)) # format사용
print_primenumber(2,10)
```

a부터 반복을 시작하여 b까지 검사를 해야하므로 a, b+1로 range를 설정해주어 반복을 실행해주었고, 2부터 ch까지 나누어지면 check를 false로 변환 및 break해주어 list에 저장해준다. 출력은 .format을 이용하여 예시대로 출력시켜주었다.
